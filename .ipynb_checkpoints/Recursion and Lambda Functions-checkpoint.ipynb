```python
def fun(x):
    if x==0:
        return 0
    x = x-1
    fun(x)
    print(x)
fun(3)    
    
```

```python
def fun(x):
    if x==0:
        return 0
    print(x)
    fun(x-1)
fun(4)    
```

```python
def fact(x):
    if x==0 or x==1:
        return 1
    return x*fact(x-1)
print(fact(5))
```

```python
def fact(n):
    if n==0 or n==1:
        return 1
    return n*fact(n-1)
print(fact(6))
```

```python
# Sum of digits in a number
# n = 456 o/p:15
def sumofDigits(n):
    if n ==0:
        return 0
    return n % 10 + sumofDigits(n//10)
print(sumofDigits(456))
print(sumofDigits(15346))
```

```python
# count digits in a number
# n = 456 o/p:15
def sumofDigits(n):
    if n ==0:
        return 0
    return 1 + sumofDigits(n//10)
print(sumofDigits(456))
print(sumofDigits(15346))
```

```python
# Find the squere of a number using lambda:
s = lambda x:x**2
print(s(10))
```

```python
# Find the squere of list of integer using map & lambda:
arr = [1,2,3,4,5]
print(list(map(lambda x:x**2, arr)))
```

```python
arr = [1,2,3,4,5]
print(list(filter(lambda x: x%2==0, arr)))
```

```python
# Checking weather the number is even or odd by using lambda:
even_or_odd = lambda x:"even" if x%2==0 else"odd"
print(even_or_odd(10))
```

```python
arr = [1,2,3,4,5]
# 1 - odd
# 2 - even
arr = [1,2,3,4,5]
print(list(map(lambda x: f"{x} - even" if x%2==0 else f"{x} - odd" , arr)))
```

```python
# Find the frequency of each character:
string = input()
count = {}
for char in string:
    if char in count:
        count[char] += 1
    else:
        count[char] = 1
print(count)        
```
