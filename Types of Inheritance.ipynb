```python
# types of inheritance:
# single inheritance
#without child class implementation
# when child class don't have constructor , in the case,parent class constructor behaves like child class constructor
class person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"my name is {self.name} and i am a {self.age} years old"
class student(person):
    pass
s = student("vijay" , 20, "male")
print(s.name)
print(s.details())
        
        
```

```python
class person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"my name is {self.name} and i am a {self.age} years old"
class student(person):
    def __init__(self,stdid,name,age,branch):
        self.name = name
        self.age = age
        self.branch = branch
        self.stdid = stdid
s = student ("vijay", 20 ,"ECE")
print(s.name)
print(s.age)
print(s.branch)
print(s.details)
```
