```python
#membership operator:
# it represts true or false (in or not in)
vijay = ["chinnnababu"]
print( "chinnababu" in vijay)
print("chinnnababu" in vijay)
```

```python
# indexing operator:
# its represent indexing valu of given list
subbulu = ["vijay" ,"kumar","kumar"]
print(subbulu[0])
print(subbulu[1])
print(subbulu[2])
           
```

```python
# slicing operator [start: stop]
king =["subbulu"]
print(king[2:4])  # its represent empty string because you wrint in sets formate[] and you give only strings in set
king =["vijay","kumar","rocky"]
print(king[1:2])
```
