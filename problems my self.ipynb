```python
#Sets Methods() : AU ,RDP,UIDS,SID
vijay = {1,2,3,4,5}
vijay.add(6) # add
print(vijay)

#update()
vijay = {1,2,3,4,5}
vijay.update([8])
print(vijay)
# remove()
v = {1,2,3,4,5}
v.remove(3)
print(v)
# discard():
vijay = {1,2,3,4,5}
vijay.update([8])
print(vijay)
# union():
vijay = {1,2,3,4,5}
raju = {6,7,8,9,0}
print(vijay.union(raju))
 # intersection():
v = {1,2,3,4,5}
r = {2,7,4,9,1}
print(v.intersection(r))
# difference():
vijay = {1,2,3,4,5}
raju = {1,2,3,4,0}
print(vijay.difference(raju))
# example 2nd
a = {1,2,3,4,5}
b = {1,2,3,4,0}
print(b.difference(a))
# symmetric_difference():
a = {1,2,3,4,5}
b = {1,2,3,4,0}
print(a.symmetric_difference(b))
# example 2:
a = {1,2,3,4,5}
b = {1,2,3,4,0}
print(b.symmetric_difference(a))
# issubset():
a = {1,2,3,4,5}
b = {1,2,3,4,0}
print(a.issubset(b))
# second example:
a = {1,2,3,4,5}
b = {1,2,3,4,5}
print(a.issubset(b)) # issubsest means is a sets elements include in b set also . its represnt True other wise False
# issupperse():
a = {1,2,3,4,5}
b = {1,2,3,4,0}
print(a.issuperset(b))
# 2nd example:
a = {1,2,3,4,5} # issuper set means is a set elements include in b set also . its represnt True other wise False
b = {1,2,3,4,5}
print(a.issubset(b))
# isdisjoint():
a = {1,2,3,4,5}
b = {1,2,3,4,0}
print(a.isdisjoint(b))
#2nd example:
a = {1,2,3,4,5}
b = {6,7,11,9,8}
print(a.isdisjoint(b))




```
