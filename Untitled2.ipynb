```python
def compute_gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print("Enter the num1 & num2 for calculating GCD")

num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))

result = compute_gcd(num1, num2)

print(f"The GCD of {num1} and {num2} is: {result}")
```
