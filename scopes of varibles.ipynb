```python
# scope of varibles
# 1. local scope
# 2. enclosing scope
# 3. global scope
# 4.Builtin scope
x = 10
def outer():
    y = 20
    def inner():
      z = 30
      print(z,y,x)
    inner()
outer()
```

```python
#updating the global scope varible value in local scope
# can't update global scope varible in local varible
x =100
def outer():
    x = x + 1
    print("Local scope varible value:" , x)
outer()
print("global scope varible vale:" , x )

    
```

```python
#updating the global scope varible value in local scope
# we can up date global in local by using keyword global
x =100
def outer():
    global x
    x = x + 1
    print("Local scope varible value:" , x)
outer()
print("global scope varible vale:" , x )
```

```python
#updating the global scope varible value in local scope
# we can up date global in local by using keyword global
x =100
def outer():
    y = 20 # Enclosing scope varible
    def inner():
        nonlocal y
        z = 30
        y = y + z # Local sope varible
        print(z, y, x) #
    inner()
outer()
```
