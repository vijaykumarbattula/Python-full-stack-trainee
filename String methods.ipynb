### String method :
Case conversion methuods :
- Upper () : It converts entire string into upper case
- lower () : It converts entire string into lower case
- title () : It converts the frist letter of every world into upper and remain lower
- swapcase () : It converts upper case letter into lower case and lower to upper
- capitalize (): It converts first character into upper case and remaing will be lower case

# Remove white space or pattern at ends of string
- strip () : By default it remove the white space at both ends of string
- lstrip () : It removes the white space at left side of string 
- rstrip () : It removes the white space at Right side of string


```python

# Split () : It splits the entire string in to lists of substring based on specified seperate 
# By default it splits at space
string= "The sky is blue and ocean is Blue "
print (string.split())
print(string.split("is"))
print(string.split('e'))
```

```python
# 'spe'.join (iterable).It join the all iterable element with seperator
words=['The' ,'sky','blue','and','occean','is','Blue']
print("".join(words))
print(" ".join(words))
print("_".join(words))
```

```python
#Substring Index Finding
#- index (substring) : It returns substring is exists in string , it returns the frist occurance of substring otherwise ValueError
#- Find (substring) : It substring is exists in string, it returns the frist occurence of substring position , otherwise -1
string = "The sky is blue and occean is Blue"
print(string.index('is'))
print(string.index('blue'))
print(string.index('and'))
print(string.find('is' , 10))
```

```python
# Count (substring): It returns the total non - over lapping occurance of substring
s="ABcBcBc"
print(s.count("B"))
print(s.count("BcB"))
lower_string = s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('z'))
```

```python
# Replace (old_substring , new substring): It replace the old .substring with new substring 
string= "The sky is blue and occean is Blue"
print(string.replace("blue" , "Red" , ))
print(string.replace('e' , 'E' ,))
print(string.replace('e' , 'E' , 2)) # is it replace the Frist 2 occurence
```
