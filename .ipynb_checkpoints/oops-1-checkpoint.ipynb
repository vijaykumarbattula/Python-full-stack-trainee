1.class : a class is blue print of a real world object or template of object
class is a collection atributes(ie. data) and method(ie. behaviour).
syntex:
     class class_name:
        #attributes
        #methods

2.object: object is instance of a class and it is a real world item
syntex:
      #object creation
       #obj = class_name(arguments)
       
       1. Attributes : the variable are defined inside the class those variables called as attributes. The attributes are used for storing data.
       Types of Attributes:
        1. class
        2.instance

```python
# define person class
class student:
    school = "XYZ school" # class attribute
    def __init__(self, name , age): # intialization constructor (magic code)
        self.name = name # Instance Attribute
        self.age = age # Instance Attribute
        print("This is student class __init__ method")
    def details(self): #
        return f" my name is {self.name} and i am a {self.age} years old"
s1 = student("raju" ,23)
s2 = student("kumar" ,24)
print(s1.name)
print(s2.name)
print(s1.details())
```

Instance Attributes :
   The variable are defined inside the instialization constructor those variables called as instance attributes.
    The instance attributes data is different to every object

2. class Attributs:
      The variables are defined out side of initilization constructor those variables called as class attributes
      The class attributes data equally shares to all class object.

```python
# define person class
class student:
    school = "kids school" # class attribute
    def __init__(my, name , age): # intialization constructor (magic code)
        my.name = name # Instance Attribute
        my.age = age # Instance Attribute
    def details(my): #
        return f" my name is {my.name} and i am a {my.age} years old"
s1 = student("raju" ,23)
s2 = student("kumar" ,24)
print(s1.name)
print(s2.name)
print(s1.school)
```
