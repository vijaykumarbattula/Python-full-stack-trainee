```python
# break statements:
list = [11,12,9,5,6,7,13]
target = 25
n = len(list)
for i in range(n):
    if list[i] == target:
        print(i)
        break
else:
    print(-1)
```

```python
# To find factors
n = int(input("Enter number: "))

print("Factors are:")
for i in range(1, n + 1):
    if n % i == 0:
        print(i)
```

```python
# To check prime numbers are not:

n = int(input("Enter number: "))

if n <= 1:
    print("Not a prime number")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```
