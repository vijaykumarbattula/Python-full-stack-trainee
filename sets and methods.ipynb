### Set methods

```python
# Element adding method 
# add(ele) :It adds the element to set if element not in set , otherwise nothing
s = {1,2,3,4,3}
print(s)
s.add(100)
s.add(5)
print(s)
```

```python
# Element removing methods
# remove(item):It removes the item if it exists in set otherwise 'KeyError'
# discard(item):It removes the item if it exists in set otherwise nothing
```

```python
s={1,2,3,4,5,}
s.remove(5)
print(s)
# s.remove(100) # KeyError
s.discard(3)
print(s)
s.discard(100) # it 
print(s)
```

```python
# sets operations
# union(set) : It returns the all unique elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
union = s1.union(s2, s3)
print(union)
res= s1| s2| s3
print(res)
```

```python
# sets operations
# intersection(set) : It returns the all common elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
intersection = s1.intersection(s2, s3)
print(intersection)
res= s1 & s2 & s3
print(res)
```

```python
# sets operations
# difference(set) : It returns the all set lements those are not in another set
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
intersection = s1.difference(s2, s3)
print(intersection)
res= s1 - s2 - s3
print(res)
print((s1 | s2) - s3)
```

```python
# sets operations
# difference(set) : It returns the non common elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
sy_diff = s1.symmetric_difference(s2)
print(sy_diff)
res = s1 ^ s2
```

```python
# issubset(set)
# issuperset(set)
# isdisjoint(set)
s1 = {1,3,5,7}
s2 = {4,5,6,7}
issubset = s1.issubset(s2 )
print(issubset)
```

```python
s1 = {1,3,5,7}
s2 = {4,5,6,7}
isdisjoint = s1.isdisjoint(s2 )
print(isdisjoint)
```

```python
s1 = {9,3,8,0}
s2 = {4,5,6,7}
issuperset = s1.issuperset(s2 )
print(issuperset)
```
