### List methods
- Item adding methods
- append(item) : It adds item at end of list
-extend(iterable): It adds the all iterator elements at end of list
-insert (pos, item): It adds item at specified position


```python
list=[1,2,3,4,5]
list.append(200)
print(list)
list.extend([10,20,30])
print(list)
```

- Item Removing Methods
- remove (item): it removes the first occurance item form the list if it is exists in list othervise ValueError
- pop(pos): it remove and return the specified position item if position is passed otherwise it remove and returns the last item from the list
- clear():It remove all elements from list 

```python
lst = [10,20,30,20,40,50,20]
val = lst.pop()
print(val)
print(lst)
val2 = lst.pop(0)
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
# index and count()
# index (item) : if iteam is present in list in returns first occurance the index otherwise value error
# count(item): It returns the count of total occurances pf item
lst = [10,20,30,20,40,50]
ind = lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# sorting items
# sort (reverse = False) : by default it sort the origina list in into Ascending order
#                       if reverser is True it sorts in Decending order
lst =[1,4,2,-5,3,-45,30]
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```
