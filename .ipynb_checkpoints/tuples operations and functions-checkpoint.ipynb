# Tuples can store different data types
# duplicates values are allowed

```python
# Indexing :
t = (10,50,80,67,49)
print(t[3])
type(t)
```

```python
# Membership ( in , not)
t = (10 , 30 , 60 , 90 )
print( 90  in  t)
```

```python
# Example 2:
t = (30 , 60 , 90 , 80)
print( 30 not in t)
type(t)
```

```python
# Functons within tuples :
# len()
t = ( 20 ,60 ,70 ,65)
print(len(t))
type(t)
```

```python
#example 2:
t = ( 89 , 90 )
print( len(t) )
type(t)
```

```python
# max ();
# we find which one is big
t = ( 90,250,780,1000)
print(max(t))
type(t)
```

```python
# Min():
# we find the smallest one in tuples
t = (890, 760, 500, 10, 5, 0)
print(min(t))
type(t)
```

```python
# sum():
t =(50 , 50, 100, 700)
print(sum(t))
type(t)
```

```python
# count() :
t = ( 3 , 6 , 9 , 9, 0)
print ( t .count ( 3 ) )
```
