```python
# 5th
products = {'pen':5, 'book':20, 'pencil':3}            
item = input()
quantity = int(input())
if item in products:
    total_amount = products.get(item) * quantity
    print(f'Total_amount of {quantity} {item} is : {total_amount}')
else:
    print(f"{item} not avaliabla")
```

```python
# Nested if
num = int(input())
if num > 0:
    if num % 2 == 0:
        print("positive even")
    else:
        print("positive odd")
else:
    if num % 2 == 0:
        print("negative even")
    else:
        print("negative odd")
```
